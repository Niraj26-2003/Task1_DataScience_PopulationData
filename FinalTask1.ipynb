```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Step 1: Load the CSV file
df = pd.read_csv('age_gender.csv' ) 
print("# Data Loaded Successfully!\n")
```

```python
print(" Preview of Dataset:")
print(df.head(11))
```

```python
print("\n Data Info:")
print(df.info())
```

```python
print("\n Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
# --- Histogram: Age Distribution ---
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=range(int(df['age'].min()), int(df['age'].max()) + 2), color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#ConclusionConclusion: Age Distribution
Most individuals are aged 25, making it the peak.

Ages 20–40 dominate, indicating a younger population.

A spike at age 0/1 may reflect infants or missing data.

Few elderly individuals (age 60+).

Age range spans 0 to 115, but higher ages are rare.

```python

# --- Bar Chart: Gender Distribution ---
gender_counts = df['gender'].value_counts().sort_index()
gender_labels = ['Male', 'Female']  

plt.figure(figsize=(6,4))
plt.bar(gender_labels[:len(gender_counts)], gender_counts, color='lightcoral', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Conclusion: Gender Distribution
The dataset contains slightly more males than females.

Males ≈ 12,500, Females ≈ 11,500 (approximate).

Overall, the gender distribution is fairly balanced.
